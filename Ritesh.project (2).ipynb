```python
# Remove duplicate rows based on the 'location' column so each location appears only once.
# Remove duplicates for restaurant_id in df3.
# Remove duplicates for user_id in df2 and df5.
# Optional: Ensure the columns used for merging are clean and consistent

Ritesh = Ritesh.drop_duplicates(subset=['location'])

Raman = Raman.drop_duplicates(subset=['restaurant_id'])

Debanjan = Debanjan.drop_duplicates(subset=['user_id'])
Jivesh = Jivesh.drop_duplicates(subset=['user_id'])


print("Unique locations in Ritesh:", Ritesh['location'].nunique())
print("Unique restaurant_ids in Raman:", Raman['restaurant_id'].nunique())
print("Unique user_ids in Debanjan:", Debanjan['user_id'].nunique())
print("Unique user_ids in Jivesh:", Jivesh['user_id'].nunique())
```

```python
# Correct merging process using 'location'.
# Optionally merge social data.
# Print the resulting DataFrame
merged1 = pd.merge(Ritesh, Raman, how='inner', on='location')  # Merge df1 with df3 on 'location'
merged2 = pd.merge(merged1, Debanjan, how='inner', on='restaurant_id')  # Merge on 'restaurant_id'
merged3 = pd.merge(merged2, Jivesh, how='inner', on='user_id')  # Merge user info on 'user_id'

final_with_social = pd.merge(merged3, Nilesh, how='outer', left_on='user_id', right_on='user_a')  # Merge social connections

print(final_with_social)
```

# 1. Top Interaction Types

```python
 # Makes the plot background white with gridlines.
 #v Sets the plot size to 10 inches wide and 6 inches tall.
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10,6)
```

```python
# Create a bar chart to show the count of each interaction type in the
# Add a title to the chart.
# Label the x-axis as "Interaction Type".
# Label the y-axis as "Count"
# Display the chart
sns.countplot(data=Debanjan, x='interaction_type', palette='Set2')
plt.title("Count of Interaction Types")
plt.xlabel("Interaction Type")
plt.ylabel("Count")
plt.show()
```

# 2 Distribution of users ages

```python
#This creates a bar chart of ages with a smooth line, 10 bars, and skyblue color.
# Add a title to the chart.
# Label the x-axis as "Age".
# Label the y-axis as "Frequency"
# Display the chart
sns.histplot(Jivesh['age'], kde=True, bins=10, color='skyblue')

plt.title("Distribution of User Ages")

plt.xlabel("Age")

plt.ylabel("Frequency")

plt.show()
```

# 3. Preferred Cuisine Popularity

```python
# Creating the count plot.
# Adding title and labels.
# Display the plot.
sns.countplot(data=Jivesh, y='preferred_cuisine', order=Jivesh['preferred_cuisine'].value_counts().index, palette='viridis')

plt.title("Preferred Cuisines")
plt.xlabel("Count")
plt.ylabel("Cuisine")

plt.show()
```

# 4. Interactions by Time of Day

```python
# Step 1: Make a copy of the data from Debanjan so the original data stays safe.
# Step 2: Add a new column to the copy called 'interaction_id', which is just a number for each row starting from 1.
# Step 3: Combine (merge) the Debanjan data with Ritesh data based on the 'interaction_id' number.
# Step 5: Add a title to the chart.
# Step 6: Label the x-axis (bottom of the chart) as "Time of Day".
# Step 7: Label the y-axis (side of the chart) as "Count" (this shows the number of interactions).
# Step 8: Show the chart

Debanjan_with_context = Debanjan.copy()

Debanjan_with_context['interaction_id'] = Debanjan_with_context.index + 1

merged = pd.merge(Debanjan_with_context, Ritesh, on='interaction_id', how='inner')

sns.countplot(data=merged, x='time_of_day', palette='pastel')

plt.title("Interactions by Time of Day")

plt.xlabel("Time of Day")

plt.ylabel("Count")

plt.show()
```

# 5. Weather vs. Interactions

```python
# Step 1: Create a bar chart to show how many interactions happened during each weather type.
# Step 2: Add a title to the chart.
# Step 3: Label the x-axis (bottom of the chart) as "Weather".
# Step 4: Label the y-axis (side of the chart) as "Count" (this shows the number of interactions).
# Step 5: Show the chart
sns.countplot(data=merged, x='weather', palette='coolwarm')

plt.title("Interactions by Weather Type")

plt.xlabel("Weather")

plt.ylabel("Count")

plt.show()
```

# 6. Top Locations by Interactions

```python
# Step 1: Create a bar chart to show interactions at each location, sorted by count.
# Step 2: Add a title to the chart.
# Step 3: Label the x-axis as "Count"
# Step 4: Label the y-axis as "Location".
# Step 5: Show the chart
sns.countplot(data=merged, y='location', order=merged['location'].value_counts().index, palette='magma')

plt.title("Top Interaction Locations")

plt.xlabel("Count")

plt.ylabel("Location")

plt.show()
```

# 7. Gender Distribution Among Users

```python
# Step 1: Create a bar chart showing the number of users by gender.
# Step 2: Add a title to the chart.
# Step 3: Label the x-axis as "Gender".
# Step 4: Label the y-axis as "Count".
# Step 5: Show the chart.
sns.countplot(data=Jivesh, x='gender', palette='Set1')

plt.title("User Gender Distribution")

plt.xlabel("Gender")

plt.ylabel("Count")

plt.show()

```

#   8. Interaction Type vs Time of Day

```python
# Step 1: Create a count plot to show interaction types by time of day, grouped by 'interaction_type'.
# Step 2: Add a title to the chart.
# Step 3: Label the x-axis as "Time of Day".
# Step 4: Label the y-axis as "Count".
# Step 5: Add a legend with the title "Interaction Type".
# Step 6: Show the chart
sns.countplot(data=merged, x='time_of_day', hue='interaction_type', palette='Accent')

plt.title("Interaction Type by Time of Day")

plt.xlabel("Time of Day")

plt.ylabel("Count")

plt.legend(title='Interaction Type')

plt.show()
```
